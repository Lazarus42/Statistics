```python
''' 
The central limit theorem (CLT) states that for i.i.d. random variables with the same expected value and finite variance the distribution of their sample average is approximately normal. 
This is a visual representation of this fact using code.
'''
```

```python
''' 
Randomly sample X_i from a Bernoulli distribution (intuitively, pretty far from Gaussian).
'''

from numpy.random import binomial
n = 1
p = .3

num_trials = 10000
bern_sample_mean_distr = []
for _ in range(10000):
    sample_mean = 0
    for _ in range(num_trials):
        sample_mean += (binomial(n, p))
    sample_mean /= num_trials
    bern_sample_mean_distr.append(sample_mean)
```

```python
import matplotlib.pyplot as plt
plt.hist(bern_sample_mean_distr, bins=50)
plt.show()
```

```python
''' 
The uniform distribution on [0,1]. Mean would be .5 and the variance would be 1/12
'''
import numpy as np
from numpy.random import uniform
low, high = 0, 1
num_trials = 10000
unif_sample_mean_distr = []
for _ in range(10000):
    unif_sample_mean_distr.append(np.mean(uniform(low, high, size = num_trials)))
```

```python
import matplotlib.pyplot as plt
plt.hist(unif_sample_mean_distr, bins=50)
plt.show()
```

```python
''' 
The exponential distribution. Mean is beta and variance is beta^2. Here lambda = 1/beta. 
'''
import numpy as np
from numpy.random import exponential
beta = 4
num_trials = 10000
exp_sample_mean_distr = []
for _ in range(10000):
    exp_sample_mean_distr.append(np.mean(exponential(beta, size = num_trials)))
```

```python
import matplotlib.pyplot as plt
plt.hist(exp_sample_mean_distr, bins=50)
plt.show()
```
